```python
cancellation_percentage = df['is_canceled'].value_counts(normalize=True)
print(cancellation_percentage)

plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not cancelled', 'Cancelled'],df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels, _ = ax1.get_legend_handles_labels()

# Adjust the legend position using bbox_to_anchor
ax1.legend(handles=legend_labels, labels=['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))

plt.title('Reservation Status in Different Hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of the reservation')
# Show the plot
plt.show()
```

```python
city_hotel = df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

Is there any price difference between City and resort hotel?

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
resort_hotel_avg = resort_hotel[['adr']].mean()
resort_hotel_avg
print(resort_hotel_avg)
city_hotel_avg = city_hotel[['adr']].mean()
print(city_hotel_avg)
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

Prices for the cancelled and reserved places 

```python

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date']) 
df['month'] = df['reservation_status_date'].dt.month

# Filter data from 2016 to 2017-09
start_date = '2016-01-01'
end_date = '2017-09-30'
filtered_df = df[(df['reservation_status_date'] >= start_date) & (df['reservation_status_date'] <= end_date)]

cancelled_data = filtered_df[filtered_df['is_canceled'] == 1]
reserved_data = filtered_df[filtered_df['is_canceled'] == 0]

cancelled_df_adr = cancelled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
reserved_df_adr = reserved_data.groupby('reservation_status_date')['adr'].mean().reset_index()

cancelled_df_adr.sort_values('reservation_status_date', inplace=True)
reserved_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate', fontsize = 20)
plt.plot(reserved_df_adr['reservation_status_date'], reserved_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.legend()
plt.xlabel('Date')
plt.ylabel('ADR')
plt.show()
```
